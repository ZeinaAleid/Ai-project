```python
df['Sleep Duration'].hist()
plt.title("Distribution of Sleep Duration")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Frequency")
plt.show()
```

**The shape of the distribution**: Bimodal

**Presence of outliers** : No outliers

```python
df['Heart Rate'].hist()
plt.title("Distribution of Heart Rate")
plt.xlabel("Heart Rate (bpm)")
plt.ylabel("Frequency")
plt.show()
```

**The shape of the distribution**:  Skewed to the right

**Presence of outliers** : There is an outlier on the right

```python
df['Stress Level'].hist()
plt.title("Distribution of Stress Level")
plt.xlabel("Stress")
plt.ylabel("Frequency")
plt.show()
```

**The shape of the distribution**: Uniform

**Presence of outliers** : No outliers

---



```python
plt.scatter(df['Age'], df['Sleep Duration'])
plt.title("Age vs Sleep Duration")
plt.xlabel("Age")
plt.ylabel("Sleep Duration")
plt.show()
```

**Is there a visible relationship?** Yes, as age increases, sleep duration slightly increases, but the relationship is not linear.

**Is the relationship strong, weak, or nonexistent?** The relationship is weak and does not strongly predict sleep duration.

```python
plt.scatter(df['Quality of Sleep'], df['Sleep Duration'])
plt.title("Quality of Sleep vs Sleep Duration")
plt.xlabel("Quality of Sleep")
plt.ylabel("Sleep Duration")
plt.show()
```

**Is there a visible relationship?** Yes, higher quality of sleep is associated with longer sleep duration.

**Is the relationship strong, weak, or nonexistent?** Strong.

```python
plt.scatter(df['Stress Level'], df['Quality of Sleep'])
plt.title("Stress Level vs Quality of Sleep")
plt.xlabel("Stress Level")
plt.ylabel("Quality of Sleep")
plt.show()
```

**Is there a visible relationship?** Yes, lower stress levels are associated with higher quality of sleep (better sleep).

**Is the relationship strong, weak, or nonexistent?** Strong and visible clearly .

```python
correlation = df[['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level']].corr()
sns.heatmap(correlation, annot=True, cmap="Reds")
plt.title("Correlation Heatmap")
plt.show()
```

**What is the strongest correlation you observed?**

Quality of Sleep & Sleep Duration: Strong positive correlation

Stress Level & Quality of Sleep: Strong negative correlation  --> The strongest one

Stress Level & Sleep Duration: Strong negative correlation

Physical Activity Level & Stress Level: Weak negative correlation
