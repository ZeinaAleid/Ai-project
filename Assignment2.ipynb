```python
import pandas as pd
```

```python
df = pd.read_csv('Sleep_health_and_lifestyle_dataset_edited.csv')
```

```python
print(df.info())
```

There is 5 categorical columns and 8 numerical columns

```python
print(df.isnull().sum())
```

There are null values in multiple columns, which vary between numerical and categorical data, such as Age, Sleep Disorder, Occupation, Quality of Sleep, BMI Category, and Heart Rate. Each column has a frequency count of null values next to its name, and the column with the highest number of null values is Sleep Disorder.

```python
print(df.isnull().any(axis=1).sum())
```

227 rows contain null values

```python
df['Age'] = df['Age'].fillna(df['Age'].median());
```

numerical column, each null value was replaced with the meadian

```python
df['Occupation'] = df['Occupation'].fillna(df['Occupation'].mode()[0])
```

categorical column, each null vaue was replaced with the first mode

```python
df['Quality of Sleep'] = df['Quality of Sleep'].fillna(df['Quality of Sleep'].mean());
```

numerical column, each null value was replaced with the mean

```python
df['BMI Category'] = df['BMI Category'].fillna(df['BMI Category'].mode()[0])
```

categorical column, each null vaue was replaced with the first mode

```python
df['Heart Rate'] = df['Heart Rate'].fillna(df['Heart Rate'].median());
```

numerical column, each null value was replaced with the meadian

```python
df['Sleep Disorder'] = df['Sleep Disorder'].fillna(df['Sleep Disorder'].mode(dropna=True)[0])
```

categorical columns, each null value was replaced with the second mode, as the first mode was NaN. I used mode(dropna=True) to exclude null values when calculating the mode.

```python
print(df.isnull().sum())
```

After filling all the null values, all columns are completely filled, and the frequency of null values is 0

```python
print(df.isnull().any(axis=1).sum())
```
